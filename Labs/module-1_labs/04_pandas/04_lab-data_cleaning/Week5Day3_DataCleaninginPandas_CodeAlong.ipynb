### Investigating and replacing null values

```python

'''
you can specify this in TWO ways: 

df.isnull(), which works if you apply it to a whole data frame! OR 
pd.isnull(df), which works better if you are applying it a specific cell, rather than the whole df (ex. pokemon challenge)

'''
df.isnull()
# output = booleans in a structure IDENTICAL to the df. If NaN, True, otherwise false.
```

```python
df.isnull().sum()
# find out how many null values

# remember these too!
# pd.isnull(df)
# pd.isnull(df['charge120'])
```

```python
# you can also check it like this:
df.isnull().sum() == 0 
```

```python
# or like this:
df.isnull().sum() > 0
```

```python
df.isnull().sum(axis=1)
# find out how many null values along axis=1 (or in the rows!)
```

```python
df.isnull().sum() / len(df)
# say for example you want to determine if the number of null values is a HIGH PERCENTAGE compared to the values in the df!
# whether you want to see whether it is a total count or a percentage depends on how you ask the Q. 
```

```python
df.count()
# note that if you take the total count, it will only show the count of values, not null values!
# so COUNT gives you the count of the NOT-MISSING values. Count all the values that are filled. 
```

```python
# BUT none of these above methods are super helpful, because you would need to go through all columns to see individuallz.
# so instead, we want a where condition to extract all columns that HAVE missing values
```

```python
# find only the columns with null values:
# subsetting! (Remember: give me all the vehicles where manufacturer = Ford.)

null_cols = df.isnull().sum()
null_cols[null_cols > 0]

# This is essentially saying:
# give me null_cols WHERE: []: null_cols > 0. 
# This is subsetting similar to SQL's WHERE command
# This is a classis subsetting task
```

```python
# instead, now I want all the null_cols that have NO missing values:
null_cols[null_cols == 0]
```

```python
# I want a list of col with > 10,000 missing values
# remember that as an analyst, if we want a dataframe with <10% missing values, we also have the percentage method above!

null_cols[null_cols > 10000]
```

```python
# let's say we see from above that column evMotor is crap and we want to drop it
# df.drop('evMotor', axis = 1)

# or give it a list of all the columns you want it to drop:
df.drop(['evMotor', 'mfrCode'], axis = 1)
```

```python
# OR: drop all of the columns with > 10000 null values (see SERIES created above)
# null_cols[null_cols > 10000]

null_cols[null_cols > 10000].index # accessing the INDEX of the series gives you all the values with 10000+ null values

cols_missing_10000 = list(null_cols[null_cols > 10000].index) # this makes a list of all the values with 10000+ null values

print(cols_missing_10000)
```

```python
# now give df.drop this new list:

df.drop(cols_missing_10000, axis = 1)

# remember, this will only be termporary. If you want it to rewrite the df, either assign it to new variable df or doing inplace=True

# note, here axis = 1 because instead of performing a function on all columns, you need to search along axis=1 (left to right) to look for the column names!!! 
# otherwise with default axis=0, it will search up and down and not find the column names. 
```

```python
# investigate and fill the remaining null values! 
```

```python
# fuel type: Elecricity! This means that they will have no cylinders and no displacement. replace by 0!
```

```python
sum(df['cylinders'].isnull())

# cylinders has 123 missing values
```

```python
df['cylinders'] = df['cylinders'].fillna(0)
```

```python
sum(df['cylinders'].isnull()) # run this again to check if it now still has missing values. Output should be 0 if successful!
```

### Cleaning Text

```python
set(df['trany'])

# look at the set (distinct values) and see if we see any issues?
# TONS of duplicates, but are just written differently, using abbreviations, spelled differently, use '-' or no '-' ETC. 
```

```python
df['trany'] = df['trany'].str.replace('-','')

set(df['trany'])

# lambda:
```

```python
# OR do multiple steps at once!
df['trany'] = df['trany'].str.replace('-','').str.replace('Automatic', 'Auto')
set(df['trany'])
```

```python
df['trany'] = df['trany'].str.replace('Automatic', 'Auto').str.replace(' ', '')
set(df['trany'])
```

```python
# what if you want to take out the brackets?
```

```python
df['trany'] = df['trany'].str.replace('()', '') # THIS WONT WORK because () are reserved operators and cant be replaced this easily. 

# what you NEED is an ESCAPE CHARACTER (or escape the special meaning of the python character):
# the backslash is an escape character in python. \ 

df['trany'] = df['trany'].str.replace('\(', '')
set(df['trany'])
```

### Dropping Duplicates

```python
df.drop_duplicates()

#note: because this gives us an output, you know it is only TEMPORARY! 
# drop_duplicates() will check the entire row to see if the entire row is duplicated! BUT this is problematic, 
# because if there is a CreatedAt column, the whole row will likely NOT be duplicated. 

# workaround is to do it on columns individually
```

```python
df[['year', 'cylinders', 'trany']].drop_duplicates()

# HERE for example, it makes sense to drop duplicates AFTER cleaning the columns. 
# because before we cleaned 'trany', we had Auto A6, Auto A-6, and Automatic A6, which now they are cleaned, will be
# DUPLICATED! SO now after cleaning and consolidating, we would drop the duplicates. 
```

```python
help(df.drop_duplicates)

# helper function tells us the default setting is to keep the FIRST occurance of the duplicate.
# helper function tells us theoretically what our options would be to change around the default option. 
```
