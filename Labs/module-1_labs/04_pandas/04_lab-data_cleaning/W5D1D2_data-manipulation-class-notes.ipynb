#### Using df.apply

```python
## df['Manufacturer'].upper() 

# error: 'Series' object has no attribute 'upper'

# when we want to apply something to each element of a LIST, we use MAP. For DFs we use APPLY!!!

df['Manufacturer'].apply(str.upper) # you need to specify str here because upper only works with strings and is a METHOD, NOT a function. 

# do it with a lambda!
# df['Manufacturer'].apply(lambda x: x.upper())
```

```python
#  df['Manufacturer'].apply()

# works with lambda functions!
# replace 'a' by 'b'

df['Manufacturer'].apply(lambda x: x.replace('a', 'b'))

# this method x.replace will come in handy for data cleaning!! You will want to use df.apply() and other methods. 

# something like: .apply(lambda x: x.replace('/', '')) will replace a slash with nothing
```

#### Dataframe Aggregation using groupby

```python
# Think about how you use GROUP BY in SQL . The logic is the same
```

```python
df.mean()
```

```python
# say we want just these descriptive stats grouped by Fuel Type
# you want one mean for each SUBGROUP (defined here by fuel type) rather than just getting one mean for the entire DF.

# df.groupby?


df.groupby(['Fuel Type']).mean()
```

```python
# you can do different aggregtion methods (mean, median, max, min, std dev, ETC!)
# you can also define MULTIPLE SUBGROUPS! 
# For each fuel type, group by manufactuer as well! Different levels of grouping.


df.groupby(['Fuel Type', 'Manufacturer']).mean()
```

```python
df.groupby(['Fuel Type'])['Cylinders'].mean()

# note that this is just df['Cylinders'].mean(), but already specifying the groupby grouping as well!
# still grouping by fuel type, but only gettings stats for cylinders. 
```

```python
df.groupby(['Fuel Type'])[['Cylinders', 'Combined MPG']].mean()
```

```python
df.groupby(['Fuel Type'])[['Cylinders', 'Combined MPG']].agg(['mean', 'max', 'min'])

# with agg() method, you can LIST all the different statistics you want to aggregate! 
```

```python
# if you want to keep your grouped df, you can RESET the index:
    
df.groupby(['Fuel Type'])[['Cylinders', 'Combined MPG']].mean().reset_index()   

# in practice, you will probably be doing this rather often! 
# like if you want KPIs by department for example.
# you will most likely be aggregating your data at a higher level ALL THE TIME, and here you have normal dfs that are easy to use in other operations! 
```
