#### Renaming columns 

```python
df.rename(columns={'Make': 'Manufacturer'}) # remember that this looks similar to a k,v pair in a dictionary. Give it the old name and the new name!
```

```python
df.head() # remember that similar to str.upper(), it reprints but does not permanently change the df!! 
```

```python
df = df.rename(columns={'Make': 'Manufacturer'}) # one way to redefine the df! 
df
```

```python
df.rename(columns={'Manufacturer': 'Make'}, inplace = True) # this "inplace" tells you to perform this operation directly in place
df
```

```python
df.rename(columns={'Make': 'Manufacturer', 'Engine Displacement': 'Displacement'}, inplace = True)
df

# Here you just do two operations at once! because of inplace = True, it is permanent, not temporary. 
```

#### Filtering Data

```python
# subsetting! equivalent to a WHERE clause in SQL. 

df['Manufacturer']=='Ford'

# you could also do df. to call the column, but it cant handle spaces in variables!!!

# this will print a ton of booleans, where False is any other make, and True is Ford
# to show only Ford, get the entire data frame where Make = Ford:

df[df['Manufacturer']=='Ford'] 
# This will also just be temp, so you need to redefine it!

df_ford = df[df['Manufacturer']=='Ford'] 
df_ford.head()
```

```python
# from the original df, only rows where manufaturer is Ford and the year is 1986

df_ford_1986 = df[(df['Manufacturer']=='Ford') & (df['Year']== 1986)] ## NOTE the error message when you had '1986' as a string, because in our df, Year is an integer and NOT a string!!
df_ford_1986.head()

# if you give any conditions, you need to put them in () to show where one condition ends and the other begins! 
# also note that BECAUSE these are conditional, they separated with "&" and NOT commas!

# will be the same error as: 
# "hello" + 5 
# and "hello" + str(5)
```

#### Calculations on Dataframes

```python
df['hello'] = 'hello' # create a new column
df.head()
```

```python
# create new column 'Age'
df['Age'] = 2020 - df['Year']
df.head() 
```

```python
df['2020'] = df['Age'] + df['Year']
df.head()

# this is just to prove the point that when you do calculations on columns, you can use MULTIPLE columns to do so. 
```
