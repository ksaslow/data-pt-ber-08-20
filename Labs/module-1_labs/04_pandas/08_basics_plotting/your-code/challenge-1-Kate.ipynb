#### Define your own data. We would like to have a variable callex which contains a list of numbers from 0 to 100. Furthermore, we would like a variables called y which is twice x, and a variable called z which is the square of x.

```python
# your code here

x = [int(x) for x in range(0, 100)]
# x = np.arange(0, 100)
y = [i*2 for i in x]
z = [i**2 for i in x]

print(z)
```

### Using DataFrames

#### In the vehicles data set that you have downloaded, use the vehicles.csv file. In this exercise we will conduct some exploratory data analysis using one plot each of scatter plot, box plot, histogram, and bar chart. 

Please provide a scatter plot between "Combined MPG" as X variable and 
"Highway MPG" as Y variable

```python
# your code here
vehicles = pd.read_csv('vehicles.csv')
vehicles.head()
```

```python
plt.scatter(vehicles['Combined MPG'],vehicles['Highway MPG'])
plt.xlabel('Combined MPG')
plt.ylabel('Highway MPG')
plt.show()
```

```python
# from solutions:

plt.scatter(x = 'Combined MPG', y = 'Highway MPG', data = vehicles)
plt.show()
```

```python
# OR (from solutions):

vehicles.plot.scatter(x= 'Combined MPG', y= 'Highway MPG')
plt.show()
```

```python
# OR (from solutions) you can plot the color as a variation of the values of a column!

vehicles.plot.scatter(x='Combined MPG', y='Highway MPG', c='Cylinders')
plt.show()
```

Bar Chart

Please provide a bar chart of the Fuel Type on the X axis and "City MPG" on the Y axis

```python
# your code here
plt.bar(x=vehicles['Fuel Type'], height=vehicles['City MPG'], color='magenta')
plt.xlabel('Fuel Type')
plt.ylabel('City MPG')
plt.show()
```

```python
# from solutions: 
grouped_fuel_type = vehicles[['Fuel Type', 'City MPG']].groupby(['Fuel Type']).mean().reset_index()
grouped_fuel_type.head()
# first step: look at what you even want to plot!
```

```python
grouped_fuel_type.plot.bar(x='Fuel Type', y='City MPG')
plt.show()
```
