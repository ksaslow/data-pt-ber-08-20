#### Define your own data. We would like to have a variable callex which contains a list of numbers from 0 to 100. Furthermore, we would like a variables called y which is twice x, and a variable called z which is the square of x.

```python
# your code here

x = [int(x) for x in range(0, 100)]
# x = np.arange(0, 100)
y = [i*2 for i in x]
z = [i**2 for i in x]

print(z)
```

#### Plot (x,y) and (x,z) on the axes in a line plot.

#### There are 2 ways of doing this - the way we have seen in class, and also using a subplot. Please research and try both!

```python
# your code here - 1st way

plt.plot(x, y, color='magenta', linewidth=4, linestyle='--')
plt.plot(x, z, color='yellow',linewidth=4, linestyle='-.')
plt.show()
```

```python
# your code here - 2nd way

# help(plt.subplot)

# plt.subplot(1,2,1) # meaning its in the first position when the 2 plots are next to each other
# plt.plot(x,y)
# plt.subplot(1,2,2) 
# plt.plot(x,z)

# plt.show()

plt.subplot(2,1,1) # meaning its in the top position when the 2 plots are above/below (note the rows and columns change)
plt.plot(x,y, color='magenta', linewidth=4)
plt.subplot(2,1,2) 
plt.plot(x,z, color='yellow', linewidth=4)

plt.show()

# note, the third argument = index, which is really telling you the POSITION
```

#### Augmenting your previous code, label your axes.

```python
# your code here
plt.plot(x, y, color='magenta', linewidth=4, linestyle='--')
plt.plot(x, z, color='yellow',linewidth=4, linestyle='-.')
plt.xlabel('X') # labeling the x axis
plt.ylabel('Y') # labeling the y axis
plt.show()
```
