Player's values are expressed in millions of euros. We would like this column to be numeric. Therefore, let's create a numeric value column. Do this by stripping all non-numeric characters from each cell. Assign this new data to `ValueNumeric`. There is no need to multiply the value to be expressed in millions. 

```python
# Your code here:

# strip the Euro sign first
fifa['ValueNumeric'] = fifa['Value'].str.replace('€', '')
fifa.head()

# Try to multiply the values to harmonize the k/m values

# This didn't make permanent changes to the df
# for x in fifa['ValueNumeric']:
#     if re.match('M', x):
#         fifa['ValueNumeric'].apply([lambda x: x * 1000000])
#     elif re.match('K', x):
#           fifa['ValueNumeric'].apply([lambda x: x * 1000])
```

```python
# from solutions:

def value_to_float(x):
    if type(x) == float or type(x) == int:
        return x
    if 'K' in x:
        if len(x) > 1:
            return float(x.replace('K', '')) * 1000
        return 1000.0
    if 'M' in x:
        if len(x) > 1:
            return float(x.replace('M', '')) * 1000000
        return 1000000.0
    
fifa['ValueNumeric'] = fifa['ValueNumeric'].apply(value_to_float)  
fifa['ValueNumeric'].value_counts()
```

# Challenge 1 - Linear Regression

Our goal with using linear regression is to create a mathematical model that will enable us to predict the outcome of one variable using one or more additional independent variables.

We'll start by ensuring there are no missing values. Examine all variables for all missing values. If there are missing values in a row, remove the entire row.

```python
# Your code here:
# check if/how many missing values
fifa.isna().sum()

# remove rows where there are nullvariables
fifa.dropna(axis=0, how='any', inplace=True)

# make sure it worked
fifa.isna().sum() 
```

Using the FIFA dataset, in the cell below, create a linear model predicting value using stamina and sprint speed. create the model using `statsmodels`. Print the model summary.

Hint: remember to add an intercept to the model using the `add_constant` function.

```python
fifa.corr()
# at first glance, neither stamina nor sprintspeed seem to be correlated with Value
```

```python
# Your code here:

sns.scatterplot(x='Stamina', y='ValueNumeric', data=fifa)
sns.scatterplot(x='SprintSpeed', y='ValueNumeric', data=fifa)
```

```python
import statsmodels.api as sm

Y = fifa['ValueNumeric']
X = fifa[['Stamina', 'SprintSpeed']]
X = sm.add_constant(X)
model = sm.OLS(Y, X)
results = model.fit()
results.summary()
```
