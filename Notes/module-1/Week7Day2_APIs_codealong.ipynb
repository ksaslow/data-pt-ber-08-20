## Working with APIs

```python
import json
import requests
import pandas as pd
```

### Simple API Requests

```python
url = 'https://jsonplaceholder.typicode.com/todos' # this placeholder website is a learning resource, APIs (in our cases) will usually look like a JSON file, NOT like a normal website!
response = requests.get(url)
results = response.json()
print(results)
```

```python
# note: GET requests here are similar to SQL queries. We will NOT be doing POST requests, which would be like an action querie in SQL, which would be able to change the data!!!
```

```python
url = 'https://cat-fact.herokuapp.com/facts'
results = requests.get(url).json()
print(results)
```

```python
pd.DataFrame(results) # if we try to turn this into a df, you will see that it isnt very pretty/helpful. 

# we call this data that is NESTED. 
# It doesn't have a flat structure, 
# but instead there will be a field with a bunch of sub-fields (hence only one column, but each row has a ton of nested data)

# we want to call the column "all"
```

```python
results.keys()

# all of the data is actually stored in this key "all"
# we therefore want to call results['all']
```

```python
pd.DataFrame(results['all'])

# from output, you can see that ['user'] is still nested!
# what you would want to do, is go into the nested column and extract each nested column to put it in a separate col.
# e.g. 'id', 'name', etc
```

```python
pd.json_normalize(results['all']) # this is a pandas METHOD that UNWRAPS the data!!!

# from output, you can see that it created new columns for the 'User' variables. now there is user._id, user.name.first, user.name.last
# it will FLATTEN the whole thing out using dots. user.id, user.name.first, etc. 

# the reason for nested fields = (quite common, especially with Big Data document databases) It allows you to have
# INCONSISTENT data points. E.g. you dont need user id, name, etc. consistently everywhere (or else missing values), NESTED FIELDS don't force you to 
# ... have identical dimensions. 

# JSON allows you to have nested fields, and nested fields allow you to have different variables in the subfields, so you need to normalize it using the method!
# WHICH for APIs is very helpful, because generally they are user-generated data, so not the most reliable. 
```
