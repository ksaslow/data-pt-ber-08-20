# Matplotlib and Seaborn

## Matplotlib refresher:

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.DataFrame()
df['x']= np.arange(0,10,1)
df['y']= df['x']**2
print(df)
```

```python
plt.plot(df['x'], df['y'])
```

```python
df.plot('x', 'y')

# NOTE: will this solve the problem I was having during the project where it didnt label the lines automatically???
```

```python
sns.lineplot(df['x'], df['y'])
```

```python
#sns.lmplot? # super helpful for regressions!
```

```python
sns.lmplot('x', 'y', data=df)
# still has the same data points as if we were just plotting the line plot (especially if using 'o' dots instead of lines!!!)
# but it also gives you a regression line, which you can switch on and off (with argument fit_reg = False, switching it OFF!)
# it also gives you the confidence around it! 
```
