```python
data = pd.read_csv('vehicles.csv')
data.head()
```

### Boxplot:
In Matplotlib

```python
data.boxplot(column = 'Combined MPG')

# you can see tht you have a ton of outliers on the higher end of the distribution.
```

```python
print(data['Combined MPG'].mean()) # higher because of the outliers.
print(data['Combined MPG'].median())
```

In Seaborn

```python
sns.boxplot(x = 'Combined MPG', data = data)
```

```python
sns.boxplot(x = 'Combined MPG', y = 'Highway MPG', data = data)

# for future, you can rotate the orientation  by 90 degrees of the labels so they are written up and down, to save space)
```

### Violin Plot

```python
sns.violinplot(x = 'Combined MPG', data = data)
```

```python
# help(sns.violinplot)
```

```python
sns.violinplot(x = 'Combined MPG', y = 'Highway MPG', data = data)
```

### Histogram

In Matplotlib

```python
plt.hist(data['Combined MPG'], bins = 50)
```

In Seaborn

```python
# sns.distplot?
```

```python
sns.distplot(data['Combined MPG'], bins = 50)

#similar to the histogram, but gives a smooth PDF (probability density function) to illustrate the data better
# just like with hist, you can specify the bin #
```

### Bar Plot

```python
# help(sns.barplot)
```

```python
sns.barplot(x = 'Highway MPG', y = 'Fuel Type', data = data)
# fuel type is the categorical variable here, hence BAR plot!
# the black bars at the end show the confidence interval! The tinier the black line, the more sure seaborn is with its estimate. The longer the bar, the less sure it is that the observation is actually where it is plotted!

# NOTE: that it automatically plots the AVERAGE highway MPG (you can change this with the argument "estimator = ")
# see below:
```

```python
sns.barplot(x = 'Highway MPG', y = 'Fuel Type', data = data, estimator = np.median)
```

```python
sns.barplot(x = 'Highway MPG', y = 'Fuel Type', data = data, saturation = 0.9)
# change the saturation to make the colors look nicer. 
```

```python
help(plt.savefig) 

# IMPORTANT: you have to call plt.savefig BEFORE plt.show! Otherwise it will save an empty white graph. 

# if you want to save the figure as a jpg, png, etc. Default is to save it within the folder you are working with. 
# give it the name, PathLike, etc. 

# the seaborn function is similar to sns.savefig but need to look it up to be sure
```

```python
fig = sns.barplot(x = 'Highway MPG', y = 'Fuel Type', data = data, saturation = 0.9)

# first you have to save the figure as fig (for example), THEN you can do fig.savefig('...')
```
